# Menampilkan Processor I9 dan Harga di Atas 150000

# Berapa jumlah laptop yang menggunakan prosesor Intel Core i7?

```python
# Menghitung jumlah laptop yang menggunakan prosesor Intel Core i7
jumlah_i7 = laptop_data[(laptop_data['Processor'] == 'Core i7') & (laptop_data["Brand"] == "MSI")].sort_values(by="Price", ascending=False)
jumlah_i7
```

# Berapa total harga laptop yang memiliki layar sentuh (Touch_Screen)?

```python
touch_screen_laptop = laptop_data[(laptop_data['Touch_Screen'] == True) & (laptop_data["Price"])].sort_values(by="Price", ascending=False)
touch_screen_laptop
```

# Berapa harga laptop dengan layar sentuh?

```python
# Memfilter DataFrame untuk laptop dengan layar sentuh
laptop_layar_sentuh = laptop_data[laptop_data["Touch_Screen"] == True].sort_values(by="Price", ascending=False)
laptop_layar_sentuh
```

# Berapa ukuran layar rata-rata dari semua laptop yang diproduksi oleh merek "HP"?

```python
# Membersihkan kolom "Screen Size" dari teks yang tidak relevan
laptop_data["Screen Size"] = laptop_data["Screen Size"].astype(str).str.extract(r"(\d+\.?\d*)").astype(float)

# Memfilter DataFrame untuk laptop HP
laptop_hp = laptop_data[laptop_data['Brand'] == "HP"]

# Menghitung rata-rata Screen Size untuk laptop HP
rata_rata_hp = laptop_hp["Screen Size"].mean()

print("Rata-rata ukuran layar untuk laptop HP adalah:", rata_rata_hp, "inci")
```
