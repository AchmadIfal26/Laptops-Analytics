# Menampilkan Processor I9 dan Harga di Atas 150000

```python
filtered_laptops = laptop_data[(laptop_data["Price"] > 150000) & (laptop_data["Processor"] == "Core i9")].sort_values(by="Price", ascending=False)
filtered_laptops
```

```python
import matplotlib.pyplot as plt

# Filter laptop dengan prosesor Core i9 dan harga di atas 150.000
filtered_laptops = laptop_data[(laptop_data["Processor"] == "Core i9") & (laptop_data["Price"] > 150000)]

# Histogram Harga
plt.figure(figsize=(10, 6))
plt.hist(filtered_laptops["Price"], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribusi Harga Laptop Core i9 dengan Harga di Atas 150000')
plt.xlabel('Harga')
plt.ylabel('Jumlah Laptop')
plt.grid(True)
plt.show()
```

# Berapa jumlah laptop yang menggunakan prosesor Intel Core i7?

```python
# Menghitung jumlah laptop yang menggunakan prosesor Intel Core i7
jumlah_i7 = laptop_data[(laptop_data['Processor'] == 'Core i7')].sort_values(by="Price", ascending=False)
jumlah_i7
```

```python
import matplotlib.pyplot as plt

# Mendapatkan data touch screen laptops
jumlah_i7 = laptop_data[(laptop_data['Processor'] == 'Core i7')].sort_values(by="Price", ascending=False)

# Menghitung jumlah laptop untuk setiap merek
jumlah_per_merek = jumlah_i7["Brand"].value_counts()

# Membuat diagram batang
plt.figure(figsize=(10, 6))
jumlah_per_merek.plot(kind='bar', color='skyblue')
plt.title('Jumlah Processor I7 Berdasarkan Merek')
plt.xlabel('Merek')
plt.ylabel('Jumlah Laptop')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Misalkan laptop_data adalah DataFrame yang sudah ada
# Filter DataFrame untuk laptop dengan layar sentuh
laptop_layar_sentuh = laptop_data[laptop_data["Touch_Screen"] == True].sort_values(by="Price", ascending=False)

# Hitung total laptop dengan layar sentuh
total_laptop_layar_sentuh = laptop_layar_sentuh.shape[0]

print(f"Total laptop dengan layar sentuh: {total_laptop_layar_sentuh}")
```

```python
import matplotlib.pyplot as plt

# Menghitung jumlah laptop dengan layar sentuh dan tidak memiliki layar sentuh
jumlah_layar_sentuh = len(laptop_data[laptop_data["Touch_Screen"] == True])
jumlah_tidak_layar_sentuh = len(laptop_data) - jumlah_layar_sentuh

# Data untuk diagram lingkaran
labels = ['True', 'False']
sizes = [jumlah_layar_sentuh, jumlah_tidak_layar_sentuh]
colors = ['skyblue', 'lightcoral']

# Membuat diagram lingkaran
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Memastikan lingkaran menjadi lingkaran
plt.title('Persentase Laptop dengan Layar Sentuh')
plt.show()
```
